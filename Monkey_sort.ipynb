```python
def is_sorted(arr):
    sorted = True
    for j in range (len(arr)-1):
        if arr[j]>arr[j+1]:
            sorted = False
    return sorted
```

```python
arr = [1,2,3,4,5,6]
is_sorted(arr)
```

```python
#Monkey Sort
import random
a = [1,2,3,4]
random.shuffle(a) # permanent shuffles the element in the array
a
```

```python
def monkey_sort(arr):
    while not is_sorted(arr):
        random.shuffle(arr)
        print(arr)
    print(arr)
    
```

```python
arr = [23,45,87,95,23,34,32,1999]
monkey_sort(arr)
```
